```python
X = heart_df.drop('target', axis=1)
y = heart_df['target']
```

#### Scaling the columns

```python
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(f"Accuracy Score on Test data: {accuracy_score(y_test, y_pred):.2f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
```

```python
class LogisticModel:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = 0
        self.lr = 0.01
        self.max_iter = 1000
        self.e = 1e-5
        self.lambda_ = 0

    def sigmoid(self, z):
        return (1 / (1 + np.exp(-z)))

    def fit(self, X, y):
        self.coef_ = np.zeros(X.shape[1])
        W_prev = np.array([-1] * (X.shape[1]))

        # Iteration counter
        i = 0
        # Run the gradient descent algorithm for a maximum number of iterations or if the norm of the gradients is below the tolerance threshold
        while i < self.max_iter and np.linalg.norm(self.coef_ - W_prev) > self.e:
            W_prev = self.coef_
            # Predict using current weights and bias
            y_cap = self.sigmoid(X @ self.coef_ + self.intercept_)

            # Calculate the gradients of the loss function with respect to the weights and bias term
            b_grad = np.sum(y_cap - y)
            W_grad = X.T @ (y_cap - y) + (self.lambda_ * self.coef_)

            # Update the weights and bias term using gradient descent
            self.coef_ = self.coef_ - self.lr * W_grad
            self.intercept_ = self.intercept_ - self.lr * b_grad

            i += 1

    def predict(self, X):
        y_pred = np.array(self.sigmoid(X @ self.coef_ +
                 self.intercept_) > 0.5, dtype=int)
        return y_pred
```

```python
model = LogisticModel()
model.fit(X_train, y_train)
```

```python
import seaborn as sns
y_pred = model.predict(X_test)
print(f"Accuracy Score on Test data: {accuracy_score(y_test, y_pred):.2f}")
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Predicted 0", "Predicted 1"],
            yticklabels=["Actual 0", "Actual 1"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
